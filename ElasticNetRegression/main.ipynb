```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('CarPrice_Assignment.csv')
dataset.head()
```

```python
i = 1
plt.figure(figsize=(15,45))
for column in dataset:
    plt.subplot(9 , 3 , i)
    plt.scatter(dataset["price"], dataset[column])
    i+=1
 
plt.show()
```

```python
dataset.corr()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
for column in dataset:
    if dataset[column].dtype == 'object':
        dataset[column] = label_encoder.fit_transform(dataset[column])
        
dataset.head()
```

```python
dataset.corr()
```

```python
from sklearn.model_selection import train_test_split
x = dataset.drop("price", axis = 1)
y = dataset['price']
x_train, x_test, y_train, y_test = train_test_split(x,y, train_size=0.8, random_state=42)
```

```python
from sklearn.linear_model import ElasticNet , LinearRegression
model = ElasticNet(alpha=1, l1_ratio=0.4)
model2 = LinearRegression()
model.fit(x_train, y_train)
model2.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred_train = model.predict(x_train)
y_pred2 = model2.predict(x_test)
y_pred_train2 = model2.predict(x_train)
```

```python
model.coef_
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
score = r2_score(y_test,y_pred)
print("R2 score is ",score)
print("Mean Squared Error is ", mean_squared_error(y_test, y_pred))
print("Mean Squared Error on train data is ", mean_squared_error(y_train, y_pred_train))
print("Mean Absolute Error is ", mean_absolute_error(y_test, y_pred))
print("Mean Absolute Percentage Error is ", mean_absolute_percentage_error(y_test, y_pred))
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_absolute_percentage_error, mean_squared_error
score = r2_score(y_test,y_pred2)
print("R2 score is ",score)
print("Mean Squared Error is ", mean_squared_error(y_test, y_pred2))
print("Mean Squared Error on train data is ", mean_squared_error(y_train, y_pred_train2))
print("Mean Absolute Error is ", mean_absolute_error(y_test, y_pred2))
print("Mean Absolute Percentage Error is ", mean_absolute_percentage_error(y_test, y_pred2))
```
